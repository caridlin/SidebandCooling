```python
import numpy as np
from scipy.optimize import curve_fit

def f(x, a, b, c):
    return c/(x + a)**2 + b

def f1(x):
    return 5.54/x**2

x = 0.1 * np.arange(300)
popt, pcov = curve_fit(f,Gamma_1, W_1, [0, 0.5, 1])
avg_fit = f(x, *popt)
x1 = 0.1 * np.arange(8, 300)
th = f1(x1)

_eta = 0.2
_omega_z = 200
m=0
N=5

import matplotlib.gridspec as gridspec
figure = plt.figure(figsize=(7, 6))
gs = gridspec.GridSpec(1, 1)
gs.update(wspace=0.3,hspace=0.4)
ax1 = plt.subplot(gs[0,0]) 
ax1.plot(Gamma_1, W_1 ,marker='.', markersize = 10, label = r'Simulation')
ax1.plot(x, avg_fit ,color = 'r',linestyle = '-', markersize = 10, label = r'Fit,$\  y = \frac{%.1f}{(x + %.1f)^2} + %.1f$' %(popt[2], popt[1], popt[0]))
ax1.plot(x1, th ,color = 'g',linestyle = '-', markersize = 10, label = r'Alt,$\  y = \frac{5.54}{x ^2} $')
#ax1.set_xlim(0,30)
#ax1.set_ylim(0,15)
ax1.set_xlabel("$\Gamma (MHz)$")
ax1.set_ylabel("Heating rate per photon ($2 E_r $)")
ax1.legend(loc = 1)
ax1.set_title('Heating rate \n'+
           '$\omega_z = 2\pi %d$ kHz, $\eta = %.2f, m=%d, N=%d$ \n' % (  _omega_z, _eta, m, N)
            + '$E_r \sim 10.46 kHz$')
ax1.grid(1)
```

$\langle \frac{dE}{dt}\rangle = R_{scatt} \frac{(\Delta F)^2}{2m \Gamma^2}$ (Alt, 2.24-2.25)

$\Delta F = m \Omega_r^2 x, \eta = \frac{kx}{\sqrt{2}}$

$\rightarrow \frac{\langle \frac{dE}{dt}\rangle}{R_{scatt} 2E_r} = \frac{(\Delta F)^2}{2m \Gamma^2 2E_r} = (\frac{m \Omega_r^2 \eta}{\hbar \Gamma  k^2})^2 = 5.54$

$\rightarrow y = \frac{5.54}{ x^2}$
