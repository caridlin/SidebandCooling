$\langle \frac{dE}{dt}\rangle = R_{scatt} \frac{(\Delta F)^2}{2m \Gamma^2}$ (Alt, 2.24-2.25)

$\Delta F = m \Omega_r^2 x, \eta = \frac{kx}{\sqrt{2}}$

$\rightarrow \frac{\langle \frac{dE}{dt}\rangle}{R_{scatt} 2E_r} = \frac{(\Delta F)^2}{2m \Gamma^2 2E_r} = (\frac{m \Omega_r^2 \eta}{\hbar \Gamma  k^2})^2 = 5.54$

$\rightarrow y = \frac{5.54}{ x^2}$
